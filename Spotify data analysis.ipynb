# Cleaning unnecessary data

```python
df= spotify_data.drop(['Unnamed: 0','Url_spotify','Uri','Url_youtube','Title'],axis=1)
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Views']=df['Views'].astype('int')
df['Likes']=df['Likes'].astype('int')
df['Comments']=df['Comments'].astype('int')
```

```python
# Collecting top 10 Artist whose music got most Views

views = df.groupby('Artist')['Views'].agg('sum').reset_index().sort_values(by='Views', ascending=False).head(10)
views
```

```python
plt.figure(figsize=(10,9))
sns.barplot(data=views,x='Artist',y='Views',hue='Artist',palette='deep')
plt.xticks(rotation=45, fontsize=8)

plt.show()
```

```python
# Collecting top 10 Liked Artist

Like=df.groupby('Artist')['Likes'].sum().reset_index(0).sort_values("Likes",ascending=False).head(10)
Like
```

```python
sns.set_style('darkgrid')
sns.barplot(data=Like,x='Likes',y='Artist',palette='magma')
plt.title("Top 10 most liked Artist",fontsize=15)
plt.show()
```

we can see  BTS and Blackpink bands got most likes from their fans. It seems South Korian bands getting more popularit day by day


```python
#Collecting top 10 Artist whose music Video got most comments

count= df.groupby('Artist')['Comments'].sum().reset_index().sort_values("Comments",ascending=False).head(10)
count
```

```python
plt.figure(figsize=(10,4))
sns.barplot(data=count,x='Artist',y='Comments',hue='Artist',palette='Set2',width=3.5)
plt.xticks(fontsize=6)
current_values= plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:.0f}'.format(x) for x in current_values])
plt.title('Top 10 artist who got most comments',fontsize=10)
plt.show()
```

```python
# Collecting top 10 songs which got maximum views, comments and likes

track= df.groupby('Track')['Views','Likes','Comments'].sum().reset_index().sort_values('Views',ascending=False).head(10)
track
```

```python
plt.pie(data=track,x=track.Views,autopct='%.2f%%',labels='Track')
plt.show()
```

We can see most viwed track in youtube is 'Swalla'. Second and third one are "Thrift shop" and "Something just like this" respectively.

```python
# Types of album available in data which got most view
Album= df.groupby('Album_type')['Views'].sum().reset_index().sort_values('Views', ascending=False)
Album
```

```python
plt.pie(data=Album,x='Views' , autopct='%.f%%', explode= (0.03,0,0),labels="Album_type")
plt.show()
```

# Correlation between our data

```python
plt.figure(figsize=(18,8))
sns.heatmap(data=df.corr(),annot=True,fmt='.1f',cmap='cubehelix')
plt.title('Heatmap')
plt.show()
```
